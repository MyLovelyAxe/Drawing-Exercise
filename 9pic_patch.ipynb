```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
```

```python
pic_folder_path = 'd:/Hard Li/Python/Drawing_Exercise_Set/ljx_pic'
pic_folder_lst = os.listdir(pic_folder_path)
print(pic_folder_lst)
```

```python
wanted = input('which pic do you want to patch?')
wanted_path = pic_folder_path + '/' + wanted
wanted_part_lst = os.listdir(wanted_path)
print(wanted_part_lst)
```

```python
sample_pic = mpimg.imread(wanted_path + '/' + wanted_part_lst[0])
H,W,C = sample_pic.shape
print(H,W,C)
```

```python
pic_lst = []
for i in wanted_part_lst:
    pic_part = mpimg.imread(wanted_path + '/' + i)
    pic_lst.append(pic_part)
print(len(pic_lst))
```

```python
pic_whole = np.zeros([H * 3,W * 3,C])
print(pic_whole.shape)
```

```python
for col in range(3):
    for row in range(3):
        pic_whole[col*W:(col+1)*W,row*H:(row+1)*H,:] = pic_lst[row+col*3]
#imshow need the input data be in range of [0,1]
#but the original pictures are usually of [0,255]
#so normalize the input date by 255 will solve the issue
#if only want to show the picture instead of save it as well
#call  ''.astype('uint8')'' will help
pic_whole = pic_whole/255
plt.imshow(pic_whole)
```

```python
mpimg.imsave(wanted_path + '/whole.jpg', arr = pic_whole)
```
