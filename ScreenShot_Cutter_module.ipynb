```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
# e.g. /home/hardli/python/HackerRank
op_path = input("Give the path where the cutting will be operated: ")
img_lst = [i for i in os.listdir(op_path) if i.endswith('png')]
print(img_lst)
```

```python
# e.g. PracticeTest_2_1.png
img_name = input("Give the name of the image to be cut: ")
img = mpimg.imread(op_path + '/' + img_name)
# screenshot size: H=1080, W=1920, C=4
H,W,C = img.shape
print(H,W)
plt.imshow(img)
```

```python
cut_img = img[130:1020,150:650,:]
plt.imshow(cut_img)
```

```python
output_path = op_path + '/' + input("Give a new name: ")
mpimg.imsave(output_path, arr = cut_img)
```
